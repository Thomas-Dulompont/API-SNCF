```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_objet = pd.read_sql_table('Objet', 'sqlite:///db.sqlite')
df_gare = pd.read_sql_table('Gare', 'sqlite:///db.sqlite')
df_meteo = pd.read_sql_table('Meteo', 'sqlite:///db.sqlite')
df_freq = pd.read_sql_table('Frequence', 'sqlite:///db.sqlite')
```

```python
df_objet.type.unique()
```

```python
df_objet['date'] = pd.to_datetime(df_objet['date'])
df_objet.set_index('date', inplace=True)
```

```python
df_week_total = df_objet.groupby([pd.Grouper(freq = 'W')]).agg(nb_objets = ('type' , 'count'))
```

```python
import plotly.express as px
import plotly.graph_objects as go

fig = go.Figure(data=[go.Histogram(nbinsx=20,x=df_week_total.nb_objets)])

# Affichage de l'histogramme
fig.show()
```

```python
df_year = df_objet.groupby([pd.Grouper(freq = 'w'), "type"]).agg(nb_objets = ('type' , 'count')).reset_index()
df_year.set_index('date', inplace=True)
```

```python
df_year
```

```python
fig = px.line(df_year, x=df_year.index, y='nb_objets', color='type', hover_name="type")
fig.show()
```
